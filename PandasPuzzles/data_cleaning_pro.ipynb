```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Setting to make numbers easier to read on display
pd.options.display.float_format = '{:20.2f}'.format

# Show all columns on output
pd.set_option('display.max_columns', 999)
```

```python
df = pd.read_csv('./data/data_to_be_cleaned.csv')
df
```

```python
df.info()
```

### Drop Null Values of client

```python
cleaned_df = df.copy()

cleaned_df.dropna(subset = ["Client"], inplace=True)

cleaned_df
```

### Drop Duplicates

```python
cleaned_df.drop_duplicates(subset=['Client', 'Beneficiary', 'AmountUSD', 'TransactionDateTime', 'ClientTier'],  inplace= True)
cleaned_df
```

### Removing of funny Characters like <, >, -

```python

cleaned_df['Client'].str.replace('[a-zA-Z0-9 ]', "", regex = True).unique()
```

```python
cleaned_df['Client'] = cleaned_df['Client'].str.strip("<>.*- ")
cleaned_df['Client'] = cleaned_df['Client'].str.replace("*", " ")

cleaned_df
```

```python
cleaned_df["Client"] =cleaned_df["Client"].str.upper()
cleaned_df["Beneficiary"] =cleaned_df["Beneficiary"].str.upper()

cleaned_df
```

```python
ax  = cleaned_df['AmountUSD'].plot(kind = 'hist', bins=10, color='skyblue', edgecolor = 'black')
ax.set_ylabel('Frequency')
ax.set_xlabel('AmountUSD')

ax.legend()

plt.show()
```

```python
mean_value = cleaned_df['AmountUSD'].mean()
median_value = cleaned_df['AmountUSD'].median()

ax  = cleaned_df['AmountUSD'].plot(kind = 'hist', bins=10, color='skyblue', edgecolor = 'black')
ax.set_ylabel('Frequency')
ax.set_xlabel('AmountUSD')

ax.axvline(mean_value, color='darkblue', linestyle='dashed', linewidth=2, label= f'Mean: {mean_value: .2f}')
ax.axvline(median_value, color='darkred', linestyle='dashed', linewidth=2, label= f'Median: {median_value: .2f}')

ax.legend()

plt.show()
```

```python
# Average Error
np.mean(np.abs(cleaned_df['AmountUSD'] - mean_value))
```

```python
# Standard Deviation

np.mean(np.abs(cleaned_df['AmountUSD'] - median_value))
```

```python
cleaned_df.fillna({"AmountUSD" : median_value}, inplace= True)
```

## Master Missing Data with Pandas: Essential techniques for Data Pros

```python
df = pd.read_csv('./data/data_to_be_cleaned2.csv')
df.head(5)
```

```python
df.info()
```

```python
df["date"] = pd.to_datetime(df["date"], format="%Y-%m-%d")
df.info()
```

```python
plt.figure(figsize = (12, 5))

plt.plot(df["date"], df["sales"], "o", label='sales', color="gold", markeredgewidth=1, markeredgecolor='black')
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("Daily Sales ($)")

plt.show()
```

```python
na_dates = df[df["sales"].isna()]["date"].values
na_dates
```

```python
plt.figure(figsize = (12, 5))

plt.plot(df["date"], df["sales"], "o", label='sales', color="gold", markeredgewidth=1, markeredgecolor='black')

for date_value in na_dates:
    plt.axvline(date_value, linestyle="dashed", color="black")
    
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("Daily Sales ($)")

plt.show()
```

```python
df["linear_interp"]= df["sales"].interpolate(method="linear")
df
```

```python
plt.figure(figsize = (12, 5))

plt.plot(df["date"], df["linear_interp"], "-", label='linear', color="red")
plt.plot(df["date"], df["sales"], "o", label='sales', color="gold", markeredgewidth=1, markeredgecolor='black')

for date_value in na_dates:
    plt.axvline(date_value, linestyle="dashed", color="black")
    
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("Daily Sales ($)")

plt.show()
```

```python
df["linear_interp"]= df["sales"].interpolate(method="linear")
df["quadratic_interp"]= df["sales"].interpolate(method="quadratic")
df["cubic_spline_interp"]= df["sales"].interpolate(method="cubicspline")
df
```

```python
plt.figure(figsize = (12, 5))

plt.plot(df["date"], df["linear_interp"], "-", label='linear', color="red")
plt.plot(df["date"], df["quadratic_interp"], "-", label='quadratic', color="blue")
plt.plot(df["date"], df["cubic_spline_interp"], "-", label='cubic_spline', color="green")
plt.plot(df["date"], df["sales"], "o", label='sales', color="gold", markeredgewidth=1, markeredgecolor='black')

for date_value in na_dates:
    plt.axvline(date_value, linestyle="dashed", color="black")
    
plt.xticks(rotation = 45)
plt.xlabel("Date")
plt.ylabel("Daily Sales ($)")

plt.show()
```

## Error Estimation

```python
## Error Estimation

errors_df = df[["date", "sales"]].copy()

errors_df["dummy_sales"] = errors_df["sales"]
errors_df.loc[(errors_df.index + 1) % 5 == 0, "dummy_sales"] = np.nan

errors_df
```

```python
errors_df.loc[:, "linear_interp"]= errors_df["dummy_sales"].interpolate(method="linear")
errors_df.loc[:, "quad_interp"]= errors_df["dummy_sales"].interpolate(method="quadratic")
errors_df.loc[:, "spline_interp"]= errors_df["dummy_sales"].interpolate(method="cubicspline")
errors_df
```

```python
errors_df["linear_error"] = np.abs(errors_df["linear_interp"] - errors_df["sales"])
errors_df["quad_error"] = np.abs(errors_df["quad_interp"] - errors_df["sales"])
errors_df["spline_error"] = np.abs(errors_df["spline_interp"] - errors_df["sales"])

errors_df
```

```python
errors_df[(~errors_df["sales"].isna()) & (errors_df["dummy_sales"].isna())]
```

```python
# Average Error

actual_errors_df = errors_df[(~errors_df["sales"].isna()) & (errors_df["dummy_sales"].isna())]

actual_errors_df[["linear_error", "quad_error", "spline_error"]].mean()
```

```python
actual_errors_df[["linear_error", "quad_error", "spline_error"]].std()
```
