```python
df = pd.read_csv("BigBasketProducts.csv", chunksize=5000)

output = pd.DataFrame()
for chunk in df:
    categories = ['category', 'brand']
    details = chunk[['product', 'category', 'brand', 'sale_price', 'market_price']]
    details.loc[:, 'count'] = 1
    summary = details.groupby(categories).sum().reset_index()
    output = pd.concat([output, summary], ignore_index = True)
```

```python
output
```

```python
final_output = output.groupby(categories).sum().reset_index()
final_output.sort_values('count')
```

```python
final_output.to_csv('aggregated-bigbasket-consumer-info.csv', index = False)
```
